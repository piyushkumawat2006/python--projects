```python
import json
```

```python
#Load the data
```

```python
def load_data(filename):
    with open(filename,"r")as f:
        data=json.load(f)

    return data
```

```python
data=load_data("store_data.json")
print(type(data))
print(data)
```

```python
#Clean and Structured the data
```

```python
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_users=set()
    for user in data:
        #print(user)
        #clean ratings-data consistency
        raw_rating=str(user["rating"]).strip().lower()
        #print(raw_rating)
        if(raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
        #print(raw_rating)
        user["rating"]=raw_rating

        #handle missing value
        raw_age=user.get("age")
        if(raw_age==None):
            user["age"]=None

        #deduplication

        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
            
```

```python
data=clean_data(data)
```

```python
#Get meaningful insights from data

def get_insight(data):
    tot_rating=0
    for user in data:
        tot_rating+=float(user["rating"])

    print(f"The avg rating is={tot_rating/len(data)}")

#percentage of  users with poor rating
    poor_rating=0
    for user in data:
        if(float(user["rating"])<3):
            poor_rating+=1

    print(f"% of user with poor rating is={(poor_rating/len(data))*100}")
        
```

```python
get_insight(data)
```

```python
#Build recommendation features

def recommendations(data):
    recommendation=[]
    for user in data:
        curr_recomment={}
        curr_recomment["name"]=user["name"]
        if(float(user["rating"])>=4):
            curr_recomment["brand"]="Apple"
        else:
            curr_recomment["brand"]="Samsung"

        recommendation.append(curr_recomment)

    return recommendation
            
```

```python
recommendations(data)
```
